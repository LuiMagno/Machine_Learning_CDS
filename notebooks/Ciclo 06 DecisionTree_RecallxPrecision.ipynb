# 0.0 Import

```python
from sklearn.datasets import make_classification
from sklearn import tree as tr
from sklearn import linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn import metrics as mt
from matplotlib import pyplot as plt
```

# 1.0 Loading Dataset

```python
# generate 2 class dataset
X, y = make_classification(n_samples=500, n_classes=2, random_state=1, n_redundant=2)

# Plotando meus dados
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# split into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=2)
```

# 2.0 Model Trainning

```python
# define
#model = tr.DecisionTreeClassifier( max_depth=2 )
model = lm.LogisticRegression(solver='lbfgs')
# fit
model.fit(X_train, y_train)

# perfomance
yhat_test = model.predict_proba(X_test)
yhat_test = yhat_test[:, 1]
```

```python
# precision, recall e tresholds
precision, recall, tresholds = mt.precision_recall_curve(y_test, yhat_test)
```

```python
plt.plot(recall, precision, marker='.', label='Model')
plt.xlabel('Recall')
plt.ylabel('Precision')
```

```python
#plt.plot(recall, precision, marker='.', label='Model')
plt.plot( tresholds, precision[:-1], 'b--', label='Precision' );
plt.plot( tresholds, recall[:-1], 'g-', label='Recall' );
plt.xlabel('Treshold')
plt.ylabel('Precision, Recall')
plt.legend()
plt.show()
```
