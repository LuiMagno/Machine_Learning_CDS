# Regressão Linear

E se eu quisesse prever a minha venda dado um determinado desconto? Será que eu consigo criar uma fórmula matemática que resuma a relação entre essas 2 variáveis, "Desconto" e "Vendas"?

- Se temos um conjunto de pontos como o mostrado abaixo, podemos traçar qualquer reta que passa por esses pontos. O eixo X representaria o desconto, e o eixo Y vendas.

- Nesse caso, vamos traçar uma reta y = x




- Podemos ver que nessa reta, sendo ela y = x, temos uma relação dela com os pontos. Mas podemos traçar infinitas retas nesse intervalo.


 

Mas qual reta melhor resume a relação entre Desconto e Vendas? É por isso que utilizamos a **Regressão Linear** para achar essa reta. OBS: Podemos ver que a reta azul e verde não são as melhores, pois estão bem longe dos 5 pontos.

**Aqui vai ficar imagem das distâncias dos pontos até a reta**

A **Regressão Linear** vai traçar essa reta de forma a minimizar a soma dos erros ao quadrado, com a técnica dos Mínimos Quadrados, utilizada em Regressão Linear.

## 1 - Importando as bibliotecas e criando o dataframe


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# Criando o dataframe abaixo
df = pd.DataFrame({
    'X': [1,2,3,4,5],
    'Y': [1.3,1.8,3.5,4,4.6]})

df.head()
```

```python
# Desenhando a figura 
fig,ax = plt.subplots()

ax.scatter(df.X, df.Y)

plt.show()
```

```python
# Adicionando a nova coluna y_reta
df['y_reta'] = df.X
df.head()
```

Essa nova reta é apenas uma previsão. Os pontos em cima dela são as minhas previsões, e a distância desses pontos pros pontos reais são os meus **erros**.

```python
# Desenhando a figura com a reta y_reta
fig,ax = plt.subplots()

ax.scatter(df.X, df.Y)
ax.plot(df.X, df.y_reta, '--r')
ax.scatter(df.X, df.y_reta)

plt.show()
```

Você vê os pontos em dourado na reta? São nossas previsões, simplesmente dizendo que Se X = 1, também Y = 1. Mas essa previsão está longe de estar correta.

Agora vamos utilizar **Machine Learning** para descobrir qual a melhor reta que resume nossos pontos nessa relação entre descontro e preço.

## 2 -  Utilizando a Regressão Linear

```python
# Instanciando nossa Regressão
regressao_linear = LinearRegression().fit(df.X.values.reshape(-1, 1), df.Y)
```

### Agora vamos ver o Coeficiente Angular e o Coeficiente Linear, para traçar a reta no nosso gráfico

```python
# Coeficiente Angular
coeficiente_angular = regressao_linear.coef_
coeficiente_angular
```

```python
# Coeficiente Linear
coeficiente_linear = regressao_linear.intercept_
coeficiente_linear
```

```python
# Agora vamos replotar nosso gráfico.

# Desenhando a figura com a reta y_reta
fig,ax = plt.subplots()

ax.scatter(df.X, df.Y)
ax.plot(df.X, df.y_reta, '--r')
ax.scatter(df.X, df.y_reta)

# Definindo a reta, com coeficiente angular e linear
x = df.X.values
y = (coeficiente_angular * x) + coeficiente_linear

# plotando a MELHOR reta
ax.plot(x, y)
ax.scatter(x, y)
plt.show()
```

```python
# Fazendo a previsão e mostrando os pontos dessa previsão
regressao_linear.predict(df.X.values.reshape(-1, 1))
```

```python
# Agora atribuindo essa regressão a uma coluna no meu dataframe
df['y_pred'] = regressao_linear.predict(df.X.values.reshape(-1, 1))
df
```

### E o erro? Vamos utilizar Mínimos Quadrados

```python
# Calculando o erro da reta vermelha (nossa primeira previsão) e da regressão
df['erro_previsao01'] = (df.Y - df.y_reta) ** 2
(df.Y - df.y_reta) ** 2
```

```python
df['erro_y_pred'] = (df.Y - df.y_pred) ** 2
(df.Y - df.y_pred) ** 2
```

### Agora vamos ver que erro é maior, somando eles


```python
df[['erro_previsao01', 'erro_y_pred']].sum()
```

```python
df[['erro_previsao01', 'erro_y_pred']].mean()
```

### Mas isso não precisa ser calculado na mão, o SkLearn já me dá funç

```python
print(mean_absolute_error(df.Y, df.y_reta))
print(mean_squared_error(df.Y, df.y_reta))

print(mean_absolute_error(df.Y, df.y_pred))
print(mean_squared_error(df.Y, df.y_pred))
```

## 3 - Utilizando a Regressão Linear numa base da dados prática

```python
# Importando o dataset prático
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
```

```python
data
```

```python
# Transformando em dataframe
df_casas = pd.DataFrame(data.data)

# Colocando as colunas
df_casas.columns = data['feature_names']
df_casas['MedHouseVal'] = data['target']

# Veja o dataframe
df_casas
```

### Agora tá na hora da diversão, Regressão Linear simples!

```python
# Vamos separar label e features
x = df_casas['MedInc'] # Minha feature
y = df_casas['MedHouseVal'] # Meu target
```

```python
# Separando em treino e teste
from sklearn.model_selection import train_test_split

# Vamos treinar com dados de treino e depois utilizados dados de teste para avaliar o modelo
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# Instanciando nossa Regressão
regressao_linear = LinearRegression().fit(X_train.values.reshape(-1, 1), y_train)
```

```python
# Fazendo a previsão
y_pred = regressao_linear.predict(X_test.values.reshape(-1, 1))

print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

### Vamos visualizar isso de forma mais gráfica

```python
# Agora vamos replotar nosso gráfico.

# Desenhando a figura com a reta y_reta
fig,ax = plt.subplots()

ax.scatter(y_pred, y_test)

plt.show()
```

### Utilizando agora a Regressão Linear Múltipla (com várias features)

```python
# Vamos selecionar as melhores features
df_casas.columns
```

```python
# Vamos separar label e features
x = df_casas[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'AveOccup', 'Latitude', 'Longitude']] # Minhas features (todas menos a target)
y = df_casas['MedHouseVal'] # Meu target
```

```python
# Separando em treino e teste
from sklearn.model_selection import train_test_split

# Vamos treinar com dados de treino e depois utilizados dados de teste para avaliar o modelo
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
# Instanciando nossa Regressão
regressao_linear = LinearRegression().fit(X_train, y_train)
```

```python
# Fazendo a previsão
y_pred = regressao_linear.predict(X_test)

print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))

# Podemos ver que diminuiu bastante
```

```python
# Agora vamos replotar nosso gráfico.

# Desenhando a figura com a reta y_reta
fig,ax = plt.subplots()

ax.scatter(y_pred, y_test)

plt.show()
```
