# 0. Imports

```python
import cv2
from sklearn import datasets as ds
from sklearn import tree as tr

from matplotlib import pyplot as plt
```

# 1.0 Load Dataset

```python
iris = ds.load_iris()

X = iris.data[:, 2:]
y = iris.target
```

```python
iris.target
```

# 2.0 Model Training

```python
# define
# max_depth: controle o tamanho máximo de quebras da árvore. Ou seja, o tamanho máximo
# de crescimento da árvore, ou a quantidade de recortes do espaço.

# outro critério - min_samples_leaf -> número mínimo de amostras que a folha deve ter 
# após a divisão

# min_samples_split: número mínimo de samples que a árvore deve ter para fazer outro 
# split

# max_features: O número máximo de atributos avaliados para a divisão de cada nó.
model_tree = tr.DecisionTreeClassifier(
    max_depth=3,
    max_features=2
)
# training (fit)
model_tree.fit(X, y)

# predict
```

```python
tr.export_graphviz(
    model_tree,
    out_file='tree.dot',
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True,
)
```
