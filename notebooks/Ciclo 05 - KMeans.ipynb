# 0.0 Imports

```python
from sklearn import datasets as ds
from matplotlib import pyplot as plt
from sklearn import cluster as ct
from sklearn import metrics as mt
import numpy as np
```

# 1.0 Loading Dataset

```python
X, y = ds.make_blobs(
    n_samples=100,
    n_features=2,
    centers=4,
    cluster_std=0.2,
    random_state=0
    )
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

# 2.0 Training Model

```python
# define
n_clusters = 4
kmeans = ct.KMeans(
            n_clusters=n_clusters,
            init='random',
            n_init=10,
            random_state=0 )

# fit (training)
labels = kmeans.fit_predict(X)

# performance (test)
silhouette_avg = mt.silhouette_score(X, labels)
print(
    "For n_clusters = {}, the average silhouette_score is : {}".format(n_clusters, silhouette_avg)
)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
```

# 3.0 Fine Tunning

```python
n_clusters = np.arange(2, 11, 1)

ss_list = []
for c in n_clusters:
# define
    kmeans = ct.KMeans(
            n_clusters=c,
            init='random',
            n_init=10,
            random_state=0 )

    # fit (training)
    labels = kmeans.fit_predict(X)

    # performance (test)
    silhouette_avg = mt.silhouette_score(X, labels)
    ss_list.append(silhouette_avg)
    print(
    "For n_clusters = {}, the average silhouette_score is : {}".format(c, silhouette_avg)
    )
```

```python
plt.plot(n_clusters, ss_list, marker='o')
plt.xlabel('Clusters K')
plt.ylabel('Silhouette Score Average')
```

# 4.0 Algorithm in Production

```python
# define
c = 4
kmeans = ct.KMeans(
            n_clusters=c,
            init='random',
            n_init=10,
            random_state=0 )

# fit (training)
labels = kmeans.fit_predict(X)

# performance (test)
silhouette_avg = mt.silhouette_score(X, labels)
ss_list.append(silhouette_avg)
print(
"For n_clusters = {}, the average silhouette_score is : {}".format(c, silhouette_avg)
)

plt.scatter(X[:, 0], X[:, 1], c=labels)
```
